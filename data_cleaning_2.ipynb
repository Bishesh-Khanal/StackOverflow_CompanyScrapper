```python
import pandas as pd
```

```python
df = pd.read_csv('companies_cleaned.csv')
```

```python
df
```

```python
del df['Unnamed: 0']
```

```python
df['Size_Average'] = df.apply(lambda x: (x.Size_least + x.Size_Max)/2 if x.Size_least != float('nan') and x.Size_Max != float('nan') else float('nan'), axis=1)
df['Size_Average'] = df['Size_Average'].apply(lambda x: str(x))
df['Size_Average'] = df['Size_Average'].apply(lambda x: float(x) if x!='nan' else float(-1))
df['Size_Average'] = df['Size_Average'].apply(lambda x: round(x) if x!=-1 else float('nan'))
```

```python
df.columns
```

```python
def remove_duplicates_comma(string):
    elements = [x.strip() for x in string.split(',')]
    unique_elements = list(set(elements))
    return ','.join(unique_elements)

def remove_duplicates_semi_column(string):
    elements = [x.strip() for x in string.split(';')]
    unique_elements = list(set(elements))
    return ';'.join(unique_elements)

df['Type'] = df['Type'].apply(remove_duplicates_comma)
df['Specialities'] = df['Specialities'].apply(remove_duplicates_comma)
df['Location'] = df['Location'].apply(remove_duplicates_semi_column)
```

```python
def make_num(df, column, num_column, separator):
    df[num_column] = df[column].apply(lambda x: len(x.split(separator)) if x != 'nan' else 0)
    
make_num(df, 'Type', 'num_Type', ',')
make_num(df, 'Specialities', 'num_Specialities', ',')
make_num(df, 'Location', 'num_Location', ';')
```

```python
df
```

```python
types = pd.Series([])
specialities = pd.Series([])
locations = pd.Series([])
def series(series_name, column, num_column, separator):
    for i in range(0, len(df)):
        for j in range(0, int(df[num_column][i])):
            item = df[column][i].split(separator)[j]
            series_name.loc[len(series_name)] = str(item)   
            
series(types, 'Type', 'num_Type', ',')
series(specialities, 'Specialities', 'num_Specialities', ',')
series(locations, 'Location', 'num_Location', ';')
```
